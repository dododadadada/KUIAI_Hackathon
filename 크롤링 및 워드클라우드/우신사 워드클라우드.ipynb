```python
import pandas as pd
casual_w = pd.read_csv('/content/drive/MyDrive/wusinsa/casual_woman.csv')
chic_w = pd.read_csv('/content/drive/MyDrive/wusinsa/chic_woman.csv')
formal_w = pd.read_csv('/content/drive/MyDrive/wusinsa/formal_woman.csv')
girlish_w = pd.read_csv('/content/drive/MyDrive/wusinsa/girlish_woman.csv')
retro_w = pd.read_csv('/content/drive/MyDrive/wusinsa/retro_woman.csv')
romantic_w = pd.read_csv('/content/drive/MyDrive/wusinsa/romantic_woman.csv')
sports_w = pd.read_csv('/content/drive/MyDrive/wusinsa/sports_woman.csv')
street_w = pd.read_csv('/content/drive/MyDrive/wusinsa/street_woman.csv')
casual_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/casual_woman2.csv')
chic_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/chic_woman2.csv')
formal_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/formal_woman2.csv')
girlish_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/girlish_woman2.csv')
retro_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/retro_woman2.csv')
romantic_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/romantic_woman2.csv')
sports_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/sports_woman2.csv')
street_w2 = pd.read_csv('/content/drive/MyDrive/wusinsa/street_woman2.csv')
```

```python
casual_woman = casual_w.loc[:,['title','hashtag']]
casual_woman['category'] = '캐주얼'
chic_woman = chic_w.loc[:,['title','hashtag']]
chic_woman['category'] = '시크'
formal_woman = formal_w.loc[:,['title','hashtag']]
formal_woman['category'] = '포멀'
girlish_woman = girlish_w.loc[:,['title','hashtag']]
girlish_woman['category'] = '걸리시'
retro_woman = retro_w.loc[:,['title','hashtag']]
retro_woman['category'] = '레트로'
romantic_woman = romantic_w.loc[:,['title','hashtag']]
romantic_woman['category'] = '로맨틱'
sports_woman = sports_w.loc[:,['title','hashtag']]
sports_woman['category'] = '스포츠'
street_woman = street_w.loc[:,['title','hashtag']]
street_woman['category'] = '스트릿'

casual_woman2 = casual_w2.loc[:,['title','hashtag']]
casual_woman2['category'] = '캐주얼'
chic_woman2 = chic_w2.loc[:,['title','hashtag']]
chic_woman2['category'] = '시크'
formal_woman2 = formal_w2.loc[:,['title','hashtag']]
formal_woman2['category'] = '포멀'
girlish_woman2 = girlish_w2.loc[:,['title','hashtag']]
girlish_woman2['category'] = '걸리시'
retro_woman2 = retro_w2.loc[:,['title','hashtag']]
retro_woman2['category'] = '레트로'
romantic_woman2 = romantic_w2.loc[:,['title','hashtag']]
romantic_woman2['category'] = '로맨틱'
sports_woman2 = sports_w2.loc[:,['title','hashtag']]
sports_woman2['category'] = '스포츠'
street_woman2 = street_w2.loc[:,['title','hashtag']]
street_woman2['category'] = '스트릿'
```

```python
hashtag_list = [casual_woman,chic_woman,formal_woman,girlish_woman,retro_woman,romantic_woman,sports_woman,street_woman,casual_woman2,chic_woman2,formal_woman2,girlish_woman2,retro_woman2,romantic_woman2,sports_woman2,street_woman2]
```

```python
##해시태그 각각을 리스트로 저장
import ast
for j in hashtag_list:
  for i in range (len(j)):
    j.at[i,"hashtag"] = ast.literal_eval(j.at[i, "hashtag"])
```

```python
wusinsa = pd.concat([casual_woman,chic_woman,formal_woman,girlish_woman,retro_woman,romantic_woman,sports_woman,street_woman,casual_woman2,chic_woman2,formal_woman2,girlish_woman2,retro_woman2,romantic_woman2,sports_woman2,street_woman2],ignore_index=True)
```

```python
import collections
```

```python
##스타일 별 사용된 해시태그를 딕셔너리 형태로 저장

##캐주얼
col = wusinsa.loc[wusinsa['category']=='캐주얼']
casual_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    casual_woman_hashlist.append(j)
casual_woman_cnt = collections.Counter(casual_woman_hashlist)

##시크
col = wusinsa.loc[wusinsa['category']=='시크']
chic_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    chic_woman_hashlist.append(j)
chic_woman_cnt = collections.Counter(chic_woman_hashlist)

##포멀
col = wusinsa.loc[wusinsa['category']=='포멀']
formal_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    formal_woman_hashlist.append(j)
formal_woman_cnt = collections.Counter(formal_woman_hashlist)

##걸리시
col = wusinsa.loc[wusinsa['category']=='걸리시']
girlish_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    girlish_woman_hashlist.append(j)
girlish_woman_cnt = collections.Counter(girlish_woman_hashlist)

##레트로
col = wusinsa.loc[wusinsa['category']=='레트로']
retro_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    retro_woman_hashlist.append(j)
retro_woman_cnt = collections.Counter(retro_woman_hashlist)

##로맨틱
col = wusinsa.loc[wusinsa['category']=='로맨틱']
romantic_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    romantic_woman_hashlist.append(j)
romantic_woman_cnt = collections.Counter(romantic_woman_hashlist)

##스포츠
col = wusinsa.loc[wusinsa['category']=='스포츠']
sports_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    sports_woman_hashlist.append(j)
sports_woman_cnt = collections.Counter(sports_woman_hashlist)

##스트릿
col = wusinsa.loc[wusinsa['category']=='스트릿']
street_woman_hashlist = []
for i in col['hashtag']:
  for j in i:
    street_woman_hashlist.append(j)
street_woman_cnt = collections.Counter(street_woman_hashlist)
```
